## Total Loan Applications 

```python
total_loan_application = df['id'].count()
print("total_loan_application:-",total_loan_application)
```

## Month to date total  loan Application (Current Month) 

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_date'].dt.year == latest_year) & (df['issue_date'].dt.month == latest_month)]
mts_loan_application = mtd_data['id'].count()

print(f"Month to date loan Application(for-{latest_issue_date.strftime('%B %Y')}):-{mts_loan_application}")
```

```python
total_funded_amount =(df['loan_amount'].sum())/1000000     # /1000000 for millions

print("total_funded_amount :-${:.2f}M".format(total_funded_amount),)
```

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_date'].dt.year == latest_year) & (df['issue_date'].dt.month == latest_month)]

mtd_total_amount = (mtd_data['loan_amount'].sum())/1000000  # in million

print("Month to date funded amount :-${:.2f}M".format(mtd_total_amount),)
```

```python
total_amount_recived = (df['total_payment'].sum())/1000000
print("Total Amount Recived:-${:.2f}M".format(total_amount_recived))
```

## Good Loan Metrics

```python
good_loan = df[df['loan_status'].isin(["Fully Paid","Current"])]

total_loan_application = df['id'].count()

good_loan_application = good_loan['id'].count()
good_loan_funded_amount = (good_loan['loan_amount'].sum())/1000000
good_loan_recived = (good_loan['total_payment'].sum())/1000000

good_loan_precentage = (good_loan_application/total_loan_application)*100

print("Good loans application:-", good_loan_application)
print("Good loans funded amount:-${:.2f}M".format(good_loan_funded_amount) )
print("Good loans recived amount :-${:.2f}M".format(good_loan_recived))
print("Precentage of good loans application:-{:.2f}%".format(good_loan_precentage))

```

## Bad loan Metrics

```python
Bad_loan = df[df['loan_status'].isin(["Charged Off"])]

total_loan_application = df['id'].count()

Bad_loan_application = Bad_loan['id'].count()
Bad_loan_funded_amount = (Bad_loan['loan_amount'].sum())/1000000
Bad_loan_recived = (Bad_loan['total_payment'].sum())/1000000

Bad_loan_precentage = (Bad_loan_application/total_loan_application)*100

print("Bad loans application:-", Bad_loan_application)
print("Bad loans funded amount:-${:.2f}M".format(Bad_loan_funded_amount) )
print("Bad loans recived amount :-${:.2f}M".format(Bad_loan_recived))
print("Precentage of Bad loans application:-{:.2f}%".format(Bad_loan_precentage))
```

# Charts
## Monthly Trends by issue date for Total funded Amount 

```python
monthly_funded = (
    df.sort_values('issue_date')
      .assign(month_name=lambda x: x['issue_date'].dt.strftime(' %b %Y'))
      .groupby('month_name',sort=False)['loan_amount'] 
      .sum()
      .div(1_000_000)
      .reset_index(name='loan_amount_millions')
)
plt.figure(figsize=(10,5))
plt.fill_between(monthly_funded['month_name'],monthly_funded['loan_amount_millions'],color='skyblue',alpha=0.5)
plt.plot(monthly_funded['month_name'],monthly_funded['loan_amount_millions'],color='blue',linewidth= 2)


for i, row in monthly_funded.iterrows():
    plt.text(i,row['loan_amount_millions']+0.1,f"{row['loan_amount_millions']:.2f}",
             ha='center',va='bottom',fontsize=9, rotation=0, color='black')
    
plt.title('Total Funded Amount by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Funded Amount($ Millions)')
plt.xticks(ticks=range(len(monthly_funded)),labels=monthly_funded['month_name'],rotation=45)
plt.grid(True,linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Monthly Trends by issue date for Total Amount Recived

```python
monthly_recived = (
    df.sort_values('issue_date')
      .assign(month_name=lambda x: x['issue_date'].dt.strftime(' %b %Y'))
      .groupby('month_name',sort=False)['total_payment'] 
      .sum()
      .div(1_000_000)
      .reset_index(name='recived_amount_millions')
)
plt.figure(figsize=(10,5))
plt.fill_between(monthly_recived['month_name'],monthly_recived['recived_amount_millions'],color='lightgreen',alpha=0.5)
plt.plot(monthly_recived['month_name'],monthly_recived['recived_amount_millions'],color='green',linewidth= 2)


for i, row in monthly_recived.iterrows():
    plt.text(i,row['recived_amount_millions']+0.1,f"{row['recived_amount_millions']:.2f}",
             ha='center',va='bottom',fontsize=9, rotation=0, color='black')
    
plt.title('Total Recived Amount by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Recived Amount($ Millions)')
plt.xticks(ticks=range(len(monthly_recived)),labels=monthly_recived['month_name'],rotation=45)
plt.grid(True,linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Monthly Trends by issue date for Total loan Application

```python
monthly_application = (
    df.sort_values('issue_date')
      .assign(month_name=lambda x: x['issue_date'].dt.strftime(' %b %Y'))
      .groupby('month_name',sort=False)['id'] 
      .count()
      .reset_index(name='loan_application_count')
)
plt.figure(figsize=(10,5))
plt.fill_between(monthly_application['month_name'],monthly_application['loan_application_count'],color='orange',alpha=0.5)
plt.plot(monthly_application['month_name'],monthly_application['loan_application_count'],color='black',linewidth= 2)


for i, row in monthly_application.iterrows():
    plt.text(i,row['loan_application_count']+0.1,f"{row['loan_application_count']}",
             ha='center',va='bottom',fontsize=9, rotation=0, color='black')
    
plt.title('Total loan application by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Number of application')
plt.xticks(ticks=range(len(monthly_application)),labels=monthly_application['month_name'],rotation=45)
plt.grid(True,linestyle='--', alpha=0.6)
```

## Regional Analysis by State for Total Funded Amount 

```python
state_funding = df.groupby('address_state')['loan_amount'].sum().sort_values(ascending=True)
state_funding_thousand = state_funding/1000

plt.figure(figsize=(10,10))
bars = plt.barh(state_funding_thousand.index,state_funding_thousand.values,color='coral')

for bar in bars:
    width = bar.get_width()
    plt.text(width+10, bar.get_y()+ bar.get_height()/2,
             f'{width:,.0f}k',va= 'center', fontsize=9)
    
plt.title('Total funded amount by state (in Thousands)')
plt.xlabel('Funded Amount ')
plt.ylabel('state')
plt.tight_layout()
plt.show()
```

## Loan term analysis by total funded amount

```python
term_funded_amount= df.groupby('term')['loan_amount'].sum()/1000000

plt.figure(figsize= (8,8))
plt.pie(
    term_funded_amount,
    labels=term_funded_amount.index,
    autopct=lambda p: f"{p:.2f}%\n${p*sum(term_funded_amount)/100:.1f}M",
    startangle=90,
    wedgeprops={'width':0.4}

)
plt.gca().add_artist(plt.Circle((0,0),0.20, color='white'))
plt.title("total funded amount by term in Millions ")
plt.show()
```

## Employee Lenght by Total Funded Amount 

```python
emp_funding = df.groupby('emp_length')['loan_amount'].sum().sort_values()/1000


plt.figure(figsize=(10,8))
bars = plt.barh(emp_funding.index,emp_funding.values,color='purple')

for bar in bars:
    width = bar.get_width()
    plt.text(width+10, bar.get_y()+ bar.get_height()/2,
             f'{width:,.0f}k',va= 'center', fontsize=9)
    
plt.title('Total Funded amount  by Employee (in Thousands)')
plt.xlabel('Funded Amount (Thousands) ')
plt.ylabel('Time in year')
plt.tight_layout()
plt.show()
```

## Loan Purpose by Total Funded Amount

```python
purpose_funding = df.groupby('purpose')['loan_amount'].sum().sort_values()/1000000


plt.figure(figsize=(10,8))
bars = plt.barh(purpose_funding.index,purpose_funding.values,color='green')

for bar in bars:
    width = bar.get_width()
    plt.text(width+10, bar.get_y()+ bar.get_height()/2,
             f'{width:,.0f}M',va= 'center', fontsize=9)
    
plt.title('Loan purpose by Total Funded  Amount ')
plt.xlabel('Funded Amount (Millions) ')
plt.ylabel('Loan Purpose')
plt.tight_layout()
plt.show()
```

## Home ownership by Total Funded Amount

```python
home_funding= df.groupby('home_ownership')['loan_amount'].sum().reset_index()
home_funding['loan_amount_millions']= home_funding['loan_amount']/1000000

fig= px.treemap(
    home_funding,
    path=['home_ownership'],
    values='loan_amount_millions',
    color= 'loan_amount_millions',
    color_continuous_scale='Blues',
    title= 'Total Funded Amount by Home Ownership(Millions)'

)
fig.show()
```
